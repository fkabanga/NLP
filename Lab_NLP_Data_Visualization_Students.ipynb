<b>N-grams</b>

<b>N-gram</b> is a contiguous sequence of n items from a given sample of text or speech.

Examples:

- â€œLatest discoveryâ€ is a 2-gram as it is a sequence of 2 words

- â€œAdvances in NLPâ€ is a 3-gram, and so on.

We would be using NLTK to view n-grams. Extract 4-grams from the piece of text below:

It's time to visualize n-grams now.
![Image](https://i1.wp.com/neptune.ai/wp-content/uploads/trigrams.png?fit=1024%2C663&ssl=1)
 The function used to view n-grams is given to you:

```python
from sklearn.feature_extraction.text import CountVectorizer #scikit learn library: https://scikit-learn.org/stable/
import seaborn as sns # seaborn is a visualization library; more here: https://seaborn.pydata.org/introduction.html#:~:text=Seaborn%20is%20a%20library%20for,explore%20and%20understand%20your%20data.
def get_top_ngram(corpus, n=None):
    vec = CountVectorizer(ngram_range=(n, n)).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0)
    words_freq = [(word, sum_words[0, idx])
                  for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:10]
```

```python
#your code here
import pandas as pd
import seaborn as sns
from nltk.util import ngrams
from collections import Counter

# Define the function to extract the top n-grams
def get_top_ngram(corpus, n, top=10):
    ngram_list = []
    for text in corpus:
        tokens = text.split()
        ngram_tokens = ngrams(tokens, n)
        ngram_list.extend(ngram_tokens)
    ngram_freq = Counter(ngram_list)
    top_ngram = ngram_freq.most_common(top)
    return top_ngram

four_grams = get_top_ngram(news_df['headline_text'], 4, top=20)

# Plot the top 20 4-grams
x, y = zip(*four_grams)
sns.barplot(x=list(y), y=[" ".join(item) for item in x])
plt.xlabel('Frequency')
plt.ylabel('4-gram')
plt.title('Top 20 4-grams in News Headlines')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
from nltk.stem import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer
from nltk import word_tokenize
def preprocess_news(df):
    corpus=[]
    stem=PorterStemmer()
    lem=WordNetLemmatizer()
    stop=set(stopwords.words('english'))
    for news in df['headline_text']:
        words=[w for w in word_tokenize(news) if (w not in stop)]

        words=[lem.lemmatize(w) for w in words if len(w)>2]

        corpus.append(words)
    return corpus

#here is the list wordlist
wordlist=preprocess_news(news)
```
